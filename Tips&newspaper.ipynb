# **newspaper project**

```python
import pandas as pd
df=pd.read_csv("/content/NewspaperData (1) (1).csv")
df
```

```python
# 'daily' is independant and 'sunday' is dependant variable. So her we perform linear regression model.
```

```python
df.shape
```

```python
df.info()
```

```python
#to find the correlation of daily and sunday edition

df.corr(numeric_only=True)
```

```python
# prompt: visualize the correlation between daily and sunday. also draw best fitting line

sns.lmplot(x='daily',y='sunday',data=df)
```

```python
import seaborn as sns
sns.distplot(df['daily'])   # distplot gives
```

```python
import statsmodels.formula.api as smf
model=smf.ols('sunday~daily',data=df).fit()
model.summary()
```

```python
model.params   #y=mx+c, c-intercept
```

```python
model.rsquared
```

```python
# ie; model have 91 percent accuracy
```

```python
model.rsquared,model.rsquared_adj
```

```python
# prompt: calculate t value for the above

model.tvalues       # tvalue is of form mx+c.
```

```python
model.pvalues   # model is ols regression model. pvalue is of form mx+c.
```

```python
# prompt: give combined dataframe for tvalue and pvalue results

df1 = pd.DataFrame(model.tvalues)
df2 = pd.DataFrame(model.pvalues)
df3 = pd.concat([df1, df2], axis=1)
df3.columns = ['tvalue', 'pvalue']
df3
```

```python
# null hypothesis is sunday edition is dependant on daily edition (if p<0.05)
# alternate hypothesis is sunday edition is independent of daily edition.
```

```python
# prompt: return the value from column=pvalue and row=daily from df3

df3['pvalue']['daily']
```

```python
alpha=0.05
if df3['pvalue']['daily']<alpha:
  print("Reject Null Hypothesis, there is a significance differance between sample mean and hypothesized population parameter")
else:
  print("Accept Null Hypothesis, there is no significance differance between sample mean and hypothesized population parameter")
```
